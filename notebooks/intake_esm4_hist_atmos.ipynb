```python
#Examples to just search for what we want from the catalog
expname_filter = ['ESM4_historical_D1']
modeling_realm = 'atmos'
model_filter = 'ESM4'
variable_id_filter = "t_ref"
ens_filter = "r1i1p1f1"
frequency = "mon"
```

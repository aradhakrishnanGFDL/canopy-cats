```python
import xarray as xr
import intake,yaml,os
import intake_esm
import matplotlib
import ast
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
col_url = "/home/a1r/tests/static-catalog.json" # github/noaa-gfdl/catalogs/statics_omonthly_opt3.json"
```

```python
col = intake.open_esm_datastore(col_url)
```

```python
expname_filter = ['CM4.5v01_om5b06_piC_noBLING']
realm = 'ocean_monthly'
frequency = "fx"
table_id = "fx"

cat_subset = col.search(experiment_id=expname_filter,
                 frequency=frequency,realm=realm,table_id = "fx")
```

```python
dset_dict = cat_subset.to_dataset_dict(cdf_kwargs={'chunks': {'time':5}, 'decode_times': False})
```

```python
dset_dict.keys()
```

```python
dset_dict['fre.CM4.5v01_om5b06_piC_noBLING.fx.fx.ocean_monthly']
```

## Get me the list of data variables in the static xarray dataset

```python
xr = dset_dict['fre.CM4.5v01_om5b06_piC_noBLING.fx.fx.ocean_monthly']
list_static_vars = list(xr.data_vars)
list_static_vars
```

```python
if "areacello" in list_static_vars:
    areacello = xr.areacello
    print(areacello)
```
